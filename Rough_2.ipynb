```
import matplotlib.pyplot as plt
```

```
x = [1, 10, 20,40]
y = [4, 19,12,28]
plt.plot(x,y)
plt.show()
```

```
x = [1, 10, 20,40]
y = [4, 19,12,28]
plt.plot(x,y,marker = '*',c = 'r')
plt.show()
```

```
x = [1, 10, 20,40]
y = [4, 19,12,28]
plt.plot(x,y, marker = 'D', linestyle = ':', c = 'b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```
x = [1, 10, 20,40]
y = [4, 19,12,28]
plt.scatter(x,y, marker = 'D', linestyle = ':', c = 'r')  # plot was for line plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```
